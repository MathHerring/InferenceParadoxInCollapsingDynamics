```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
from matplotlib import ticker
import seaborn as sns
%matplotlib inline
```

```python

sns.set()
sns.set_style("white")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.0})
```

# Plotting the boundary classification

```python
cmap = mpl.colormaps['RdBu_r']
lwdth = 3.5

fig, ax = plt.subplots(1,2, figsize=(8.5,4.5))

#### fill areas
x = np.arange(-2, 3.5, 0.01)
def y(x):
    # exit boundary gamma > 0
    ys = []
    for x1 in x:
        if x1 < 1:
            ys.append(x1-1)
        elif x1 < 2:
            ys.append(4)
        else:
            ys.append(1)
    return ys
ax[0].fill_between(x, y(x), -3, color=cmap(0.15), alpha=0.6) # exit 

x2 = np.arange(-2, 1.0, 0.01)
y2 = x2-1
ax[0].fill_between(x2, y2, 4, color=cmap(0.25), alpha=0.6) # regular

x3 = np.arange(2, 4, 0.01)
ax[0].fill_between(x3, 1,3 , color=cmap(0.95), alpha=0.6) # natural


#### lines
ax[0].plot([1,1], [0,3], '--', color = cmap(0.15), linewidth=lwdth) # beta = 1
ax[0].plot([-0.5,1], [-1.5,0], '--', color = cmap(0.15), linewidth=lwdth) # alpha - beta = -1
ax[0].plot([-1,-0.5], [-2,-1.5],'--', color = cmap(0.25), linewidth=lwdth) # alpha - beta = -1
ax[0].plot([2,2], [1,4],'--', color = cmap(0.95), linewidth=lwdth) # alpha - beta = -1 #natural
ax[0].plot([2,4], [1,1],'--', color = cmap(0.95), linewidth=lwdth) # alpha - beta = -1 #naturla

ax[0].plot([-0.5,0.1], [-1.5,-1.5], '--', color = 'k', linewidth=1) # alpha - beta = -1
ax[0].text(0.15, -1.58, r'$\alpha = -\frac{2\gamma}{D}$', fontsize=16)





### naming boundaries
ax[0].text(2.1, 1.5, r'Natural', fontsize=20)
ax[0].text(1, -1, r'Exit', fontsize=20)
ax[0].text(-1.1, 0.5, r'Regular', fontsize=20)
#ax[0].text(1.5, 0.5, r'Exit', fontsize=20)

#### labelling
ax[0].text(3.3, -0.05, r'$\beta$', fontsize=14)
ax[0].text(-0.07, 2.6, r'$\alpha$', fontsize=14)
ax[0].set_title(r'$\gamma > 0$', fontsize=20)

#### formatting axis
ax[0].set_xlim([-1.5,3.2])
ax[0].set_ylim([-2,2.5])
ax[0].spines['left'].set_position('zero')
ax[0].spines['bottom'].set_position('zero')
ax[0].spines['right'].set_color('none')
ax[0].spines['top'].set_color('none')
ax[0].xaxis.set_ticks_position('bottom')
ax[0].yaxis.set_ticks_position('left')
ax[0].xaxis.set_major_locator(MaxNLocator(integer=True))
ax[0].yaxis.set_major_locator(MaxNLocator(integer=True))
xticks = ax[0].xaxis.get_major_ticks()
xticks[2].set_visible(False)
yticks = ax[0].yaxis.get_major_ticks()
yticks[2].set_visible(False)




###################### plot 2 \gamma > 0

#### fill areas
x = np.arange(-2, 3.5, 0.01)
def y(x):
    # exit boundary gamma > 0
    ys = []
    for x1 in x:
        if x1 < 2:
            ys.append(x1-1)
        else:
            ys.append(1)
    return ys
ax[1].fill_between(x, y(x), -3, color=cmap(0.8), alpha=0.6) # entrance

x2 = np.arange(-3, 1.0, 0.01)
y2 = x2-1
ax[1].fill_between(x2, y2, 4, color=cmap(0.25), alpha=0.6) # regular

x3 = np.arange(2, 4, 0.01)
ax[1].fill_between(x3, 1,3 , color=cmap(0.95), alpha=0.6) # natural

x4 = np.arange(1.0, 2.0, 0.01)
def y4(x):
    # exit boundary gamma > 0
    ys = []
    for x1 in x:
        if x1 < 2:
            ys.append(4)
        else:
            ys.append(1)
    return ys
ax[1].fill_between(x4, y4(x4), x4-1, color=cmap(0.15), alpha=0.6) # exit

#### lines
ax[1].plot([1,1], [0,3], '--', color = cmap(0.15), linewidth=lwdth) # beta = 1
#ax[1].plot([-0.5,1], [-1.5,0], '--', color = cmap(0.15), linewidth=2) # alpha - beta = -1
ax[1].plot([-1,1], [-2,0],'--', color = cmap(0.25), linewidth=lwdth) # alpha - beta = -1
ax[1].plot([2,2], [1,4],'--', color = cmap(0.95), linewidth=lwdth) # alpha - beta = -1 #natural
ax[1].plot([2,4], [1,1],'--', color = cmap(0.95), linewidth=lwdth) # alpha - beta = -1 #natural
ax[1].plot([1.5,2], [0.5,1],'--', color = cmap(0.15), linewidth=lwdth) # alpha - beta = -1 #exit
ax[1].plot([1,1.5], [0,0.5],'--', color = cmap(0.25), linewidth=lwdth) # alpha - beta = -1 #regular
ax[1].plot([-1,2], [-2.1,0.9],'--', color = cmap(0.8), linewidth=lwdth) # alpha - beta = -1 #entrance


ax[1].plot([1.5,1.85], [0.5,0.5], '--', color = 'k', linewidth=1) # alpha - beta = -1
ax[1].text(1.89, 0.43, r'$\alpha = -\frac{2\gamma}{D}$', fontsize=16)
#ax[1].plot([1.2,1.85], [0.15,0.15], '--', color = 'k', linewidth=1) # alpha - beta = -1
#ax[1].text(1.89, 0.03, r'$\gamma < -\frac{D}{2}$', fontsize=16)
ax[1].plot([-0.4,0.1], [-1.5,-1.5], '--', color = 'k', linewidth=1) # alpha - beta = -1
ax[1].text(0.15, -1.58, r'$\gamma < -\frac{D}{2}$', fontsize=16)


### naming boundaries
ax[1].text(2.1, 1.5, r'Natural', fontsize=20)
ax[1].text(1, -1, r'Entrance', fontsize=20)
ax[1].text(-1.1, 0.5, r'Regular', fontsize=20)
ax[1].text(1.1, 1.0, r'Exit', fontsize=20)

# labelling
ax[1].text(3.3, -0.05, r'$\beta$', fontsize=14)
ax[1].text(-0.07, 2.6, r'$\alpha$', fontsize=14)
ax[1].set_title(r'$\gamma < 0$', fontsize=20)

# formatting axis
ax[1].set_xlim([-1.5,3.2])
ax[1].set_ylim([-2,2.5])
ax[1].spines['left'].set_position('zero')
ax[1].spines['bottom'].set_position('zero')
ax[1].spines['right'].set_color('none')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].yaxis.set_ticks_position('left')
ax[1].xaxis.set_major_locator(MaxNLocator(integer=True))
ax[1].yaxis.set_major_locator(MaxNLocator(integer=True))
xticks = ax[1].xaxis.get_major_ticks()
xticks[2].set_visible(False)
yticks = ax[1].yaxis.get_major_ticks()
yticks[2].set_visible(False)

ax[0].tick_params(axis='both', which='major', labelsize=14)
ax[1].tick_params(axis='both', which='major', labelsize=14)

plt.tight_layout()
plt.savefig('boundary_final_2.pdf')
plt.show()
```
