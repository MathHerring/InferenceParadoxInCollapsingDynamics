#### some theory results

```python
def theory_mean_bessel(gam, D, x):
    return np.sqrt(2*D*x) * sps.gamma(gam/D +2)/sps.gamma(gam/D +3/2)
def theory_var_bessel(gam, D, x):
    return D * x * (3 + 2*gam/D - (2*sps.gamma(2 + gam/D)**2)/sps.gamma(3/2 + gam/D)**2)
def cov_bessel_single(gam, D, s, t):
    #s = np.min(t1, t2)
    #t = np.max(t1, t2)
    cov = (2*D*sps.gamma(2 + gam/D)**2 *(-np.sqrt(s*t) + np.sqrt(s) * t**(-2 - gam/D) * (-s + t)**(5/2 + gam/D)* sps.hyp2f1(2 + gam/D, 2 + gam/D, 3/2 + gam/D, s/t)))/sps.gamma(3/2 + gam/D)**2
    return cov 

def theory_corr_single(gam, D, alpha, s, t): 
    #s = np.min(t1, t2)
    #t = np.max(t1, t2)
    return (s/t)**((3/2) + 1/(alpha-1))

def theory_cov(gam, D, alpha, t):
    var = D * (1 - alpha) / (1 - 3 * alpha) * t
    
    N = len(t)
    cov_array = np.zeros((N, N))
    
    for i in np.arange(N-1):
        s = t[i]
        s2 = t[i+1:]  # Elements after s
        
        cov_row = theory_corr_single(gam, D, alpha, s, s2) * np.sqrt(var[i]) * np.sqrt(var[i+1:])
        cov_array[i+1:, i] = cov_row
        cov_array[i, i+1:] = cov_row
    np.fill_diagonal(cov_array, var)
    
    return cov_array

################################################### WITH BETA


def theory_cov_single_beta(gam, D, alpha, beta, s, t): 
    #s = np.min(t1, t2)
    #t = np.max(t1, t2)
    cov_single_nom = D * ((1-alpha)*gam*t)**(-alpha/(alpha-1)) * ((1-alpha)*gam*s)**((-2*alpha+beta+1)/(1-alpha))
    cov_single_denom = gam*(1-3*alpha+beta)
    
    return cov_single_nom/cov_single_denom

def theory_cov_beta(gam, D, alpha, beta, t):
    var = D /gam * ((1 - alpha) * gam * t)**((1-alpha+beta)/(1-alpha)) / (1 - 3 * alpha + beta) 
    
    N = len(t)
    cov_array = np.zeros((N, N))
    
    for i in np.arange(N-1):
        s = t[i]
        s2 = t[i+1:]  # Elements after s
        
        cov_row = theory_cov_single_beta(gam, D, alpha, beta, s, s2)
        cov_array[i+1:, i] = cov_row
        cov_array[i, i+1:] = cov_row
    np.fill_diagonal(cov_array, var)
    
    return cov_array


############################################### BESSEL

def cov_bessel_single_tup(gam, D, tup):
    #s = np.min(t1, t2)
    #t = np.max(t1, t2)
    s = tup[0]
    t = tup[1]
    cov = (2*D*sps.gamma(2 + gam/D)**2 *(-np.sqrt(s*t) + np.sqrt(s) * t**(-2 - gam/D) * (-s + t)**(5/2 + gam/D)* sps.hyp2f1(2 + gam/D, 2 + gam/D, 3/2 + gam/D, s/t)))/sps.gamma(3/2 + gam/D)**2
    return cov 

def theory_cov_bessel(gam, D, t):
    
    N = len(t)
    cov_array = np.zeros((N,N))
    
    for i in np.arange(N-1):
        s = t[i]
        s2 = t[i+1:]
        
        cov_row = cov_bessel_single(gam, D, s, s2)
        cov_array[i+1:, i] = cov_row
        cov_array[i, i+1:] = cov_row
        
        np.fill_diagonal(cov_array, theory_var_bessel(gam, D, t))
    
    return cov_array
    
```

### entropy function test

```python
def entropy_from_eigen(evs):
    k = len(evs)
    norm = 0.5*k*np.log2(2*np.pi*np.exp(1))
    ent = 0.5*np.log2(np.prod(evs))
    
    return norm + ent
```

```python
t = np.linspace(0,3,100)

entropies = []
alphas = []
for alpha in np.linspace(-20,0,200):
    print(alpha)
    
    cov_purenoise = theory_cov(1, 0.2, alpha, t)
    try:
        eigen, eigenvec = np.linalg.eig(cov_purenoise)   

        entr = entropy_from_eigen(eigen[:10])

        entropies.append(entr)
        alphas.append(alpha)
    except:
        continue
```

### entropy theory plotting

```python
### beta = -1

from scipy.special import gamma, hyp2f1

# Set the d and gamma values based on your requirement.
d_value = 1
gamma_param_value = 1 

# Create an array of 10 values, evenly spaced from 0 to 3.
tau = 3
num_values = 100
s_values = np.linspace(0, tau, num_values)
t_values = np.linspace(0, tau, num_values)

# Function to calculate covariance
def cov_function_min1(d, s, t, gamma_param):
    min_value = min(s/t, t/s) if t!=0 and s!=0 else 0
    return (3 * np.power(3, 1/3) * np.power(d*d*s*t, 1/3) * np.power(sp.special.gamma(5/3 + (2 * gamma_param) / (3 * d)), 2) *
            (-1 + hyp2f1(-1/3, -1/3, 2/3 * (2 + gamma_param / d), min_value))) / (np.power(2, 2/3) *
                                                                                   np.power(sp.special.gamma(2/3 * (2 + gamma_param / d)), 2))

# Create matrix
matrix_values = [[cov_function_min1(d_value, s, t, gamma_param_value) for t in t_values] for s in s_values]
cov_matrix = np.array(matrix_values)

# Plot matrix
plt.imshow(cov_matrix, cmap='RdBu_r', origin='lower')
plt.colorbar()
plt.show()
```

```python
#### beta = 1/2
# Set the d and gamma values based on your requirement.
d_value = 1
gamma_param_value = 1 

# Create an array of 10 values, evenly spaced from 0 to 3.
tau = 3
num_values = 100
s_values = np.linspace(0, tau, num_values)
t_values = np.linspace(0, tau, num_values)

# Function to calculate covariance
def cov_function_05(d, s, t, gamma_param):
    min_value = min(s/t, t/s) if t!=0 and s!=0 else 0
    return (-1 / (16 * np.power(sp.special.gamma(5/3 + (4 * gamma_param) / (3 * d)), 2)) * 
            9 * np.power(3, 2/3) *
            np.power(sp.special.gamma(7/3 + (4 * gamma_param) / (3 * d)), 2) * 
            ((np.power(d*d*s*t, 2/3)) -
            np.power(d, 4/3) * np.power(s*t, 2/3) *
            hyp2f1(-2/3, -2/3, 5/3 + (4 * gamma_param) / (3*d), min_value)))

# Create matrix
matrix_values = [[cov_function_05(d_value, s, t, gamma_param_value) for t in t_values] for s in s_values]
cov_matrix = np.array(matrix_values)

# Plot matrix
plt.imshow(cov_matrix, cmap='seismic', origin='lower')
plt.colorbar()
plt.show()
```

summary of theory functions

```python
def theory_cov_single_1(gam, D, s, t): 
    # alpha= 0, beta=1
    #s = np.min(t1, t2)
    #t = np.max(t1, t2)
    cov_single = 0.5 * D * s**2 *(D+gam)
    return cov_single

def theory_cov_1(gam, D, t):
    var = theory_cov_single_1(gam, D, t, t)
    
    N = len(t)
    cov_array = np.zeros((N, N))
    
    for i in np.arange(N-1):
        s = t[i]
        s2 = t[i+1:]  # Elements after s
        
        cov_row = theory_cov_single_1(gam, D, s, s2)
        cov_array[i+1:, i] = cov_row
        cov_array[i, i+1:] = cov_row
    np.fill_diagonal(cov_array, var)
    
    return cov_array
```

```python
def entrtheo_1(d, gamma):
    tau = 0.1
    num_values = 100
    t_values = np.linspace(0, tau, num_values)
    s_values = np.linspace(0, tau, num_values)
    
    matrix_values = [[cov_function_1(d, s, t, gamma) for t in t_values] for s in s_values]
    cov_matrix = np.array(matrix_values)
    
    # entropy
    eigs = np.linalg.eig(cov_matrix)    
    eigen = eigs.eigenvalues

    entr = entropy_from_eigen(eigen[:10])
    return entr
    
def entrtheo_05(d, gamma):
    tau = 0.1
    num_values = 100
    t_values = np.linspace(0, tau, num_values)
    s_values = np.linspace(0, tau, num_values)
    
    matrix_values = [[cov_function_05(d, s, t, gamma) for t in t_values] for s in s_values]
    cov_matrix = np.array(matrix_values)
    
    # entropy
    eigs = np.linalg.eig(cov_matrix)    
    eigen = eigs.eigenvalues

    entr = entropy_from_eigen(eigen[:10])
    return entr

def entrtheo_0(d,gamma):
    tau = 0.1
    num_values = 100
    t_values = np.linspace(0, tau, num_values)
    s_values = np.linspace(0, tau, num_values)
    
    matrix_values = [[cov_function_0(d, s, t, gamma) for t in t_values] for s in s_values]
    cov_matrix = np.array(matrix_values)
    
    # entropy
    eigs = np.linalg.eig(cov_matrix)    
    eigen = eigs.eigenvalues

    entr = entropy_from_eigen(eigen[:10])
    return entr

def entrtheo_min1(d, gamma):
    tau = 0.1
    num_values = 100
    t_values = np.linspace(0, tau, num_values)
    s_values = np.linspace(0, tau, num_values)
    
    matrix_values = [[cov_function_min1(d, s, t, gamma) for t in t_values] for s in s_values]
    cov_matrix = np.array(matrix_values)
    
    # entropy
    eigs = np.linalg.eig(cov_matrix)    
    eigen = eigs.eigenvalues

    entr = entropy_from_eigen(eigen[:10])
    return entr
```
