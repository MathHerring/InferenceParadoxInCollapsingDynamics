```python
# basics
import numpy as np
import scipy.special as sp
import os

# plotting
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
sns.set_style("white")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.0})

# seed the random number generator
rstate = np.random.default_rng(5647)
```

# plot cluster results

**variable inference interval**

```python
evslen = np.loadtxt('evidence_opaque_largerinterval_CIR.csv')
evslen_vas = np.loadtxt('evidence_opaque_largerinterval_vasicek.csv')
evslen_bessel = np.loadtxt('evidence_opaque_largerinterval_bessel.csv')
evslen_WF = np.loadtxt('evidence_opaque_largerinterval_WF.csv')
```

```python
t_lengths = np.arange(0.01, 2.01, 0.01)

t_wf = np.zeros(len(t_lengths)+1)
evwf = np.copy(t_wf)
t_wf[1:] = t_lengths
evwf[1:] = evslen_WF

plt.figure(figsize=(6,5.5))

plt.plot(t_lengths, evslen, color='b', label='Cox-Ingersoll-Ross')
plt.plot(t_lengths, evslen_vas, color='purple', label='Vasicek')
plt.plot(t_wf, evwf, color='orange', label='Wright-Fisher')
plt.plot(t_lengths, evslen_bessel, color='g', label='Bessel')

plt.fill_between([0.0,0.28], -6,6000 , color='gray', alpha=0.4) 

plt.plot([0,4], [0,0], c='k', linewidth=1)


plt.ylabel(r'$\Delta$  log$\mathcal{Z}$')
plt.xlabel(r'inference window size $\tau_{inf}$')


plt.ylim(-5,5080)
plt.xlim(0,.8)

plt.title(r'$\tau_0 = 0.1$')

plt.legend()


plt.tight_layout()

plt.savefig('opaque_inference_comparison_variablewindowsize_inset.pdf')

plt.show()
```
