# mean and variance

implementing analytical solutions

```python
def mean_theory(y0, d, t, gamma):
    term1 = 2 ** (1 + (2 * gamma) / d)
    term2 = np.exp(-((2 * y0) / (d * t)))
    term3 = ((d * t) / y0) ** (-((2 * gamma) / d))
    term4 = y0 * (d * t + 2 * y0)
    term5 = d ** 2 * t ** 2 * (sp.gamma(3 + (2 * gamma) / d) - sp.gamma(3 + (2 * gamma) / d) * sp.gammaincc(3 + (2 * gamma) / d, (2 * y0) / (d * t)))
    term6 = 2 * t * (d + 2 * gamma) * (sp.gamma(1 + (2 * gamma) / d) - sp.gamma(1 + (2 * gamma) / d) * sp.gammaincc(1 + (2 * gamma) / d, (2 * y0) / (d * t)))
    return (term1 * term2 * term3 * term4 + term5) / term6

def var_theory(y0, d, t, gamma):
    term1 = -(2 ** (1 + (2 * gamma) / d) * np.exp(-((2 * y0) / (d * t))) * ((d * t) / y0) ** (-((2 * gamma) / d)) * y0 * (d * t + 2 * y0) + d ** 2 * t ** 2 * (sp.gamma(3 + (2 * gamma) / d) - sp.gamma(3 + (2 * gamma) / d) * sp.gammaincc(3 + (2 * gamma) / d, (2 * y0) / (d * t)))) ** 2
    term2 = t * (d + 2 * gamma) * (sp.gamma(1 + (2 * gamma) / d) - sp.gamma(1 + (2 * gamma) / d) * sp.gammaincc(1 + (2 * gamma) / d, (2 * y0) / (d * t)))
    term3 = 4 ** ((d + gamma) / d) * np.exp(-((2 * y0) / (d * t))) * ((d * t) / y0) ** (-((2 * gamma) / d)) * y0 * (d ** 2 * t ** 2 + 2 * d * t * y0 + 2 * y0 ** 2) + d ** 3 * t ** 3 * (sp.gamma(4 + (2 * gamma) / d) - sp.gamma(4 + (2 * gamma) / d) * sp.gammaincc(4 + (2 * gamma) / d, (2 * y0) / (d * t)))
    term4 = 4 * t ** 2 * (d + 2 * gamma) ** 2 * (sp.gamma(1 + (2 * gamma) / d) - sp.gamma(1 + (2 * gamma) / d) * sp.gammaincc(1 + (2 * gamma) / d, (2 * y0) / (d * t))) ** 2
    return (term1 + term2 * term3)/ term4
    #return 1/ term4
```

```python
def mean_bessel_theory(alpha, d, t, gamma_val):
    term1 = 2**(1/(-1 + alpha)) * (d*t)**(1/(1 - alpha)) * (1 - alpha)**(-(2/(-1 + alpha)))
    term2 = sp.gamma((3 * d - d * alpha + 2 * gamma_val)/(d - d * alpha))
    term3 = sp.gamma((2 * d - d * alpha + 2 * gamma_val)/(d - d * alpha))
    return term1 * term2 / term3

def var_bessel_theory(alpha, d, t, gamma_val):
    term1 = 4**(1/(-1 + alpha)) * (d * t)**(-(2/(-1 + alpha))) * (1 - alpha)**(-(4/(-1 + alpha)))
    term2 = sp.gamma((3 * d - d * alpha + 2 * gamma_val)/(d - d * alpha))**2
    term3 = sp.gamma((2 * d - d * alpha + 2 * gamma_val)/(d - d * alpha)) * sp.gamma((4 * d - d * alpha + 2 * gamma_val)/(d - d * alpha))
    return term1 * (-term2 + term3) / sp.gamma((2 * d - d * alpha + 2 * gamma_val)/(d - d * alpha))**2
```

# trajectories and cumulative

```python
def cum_theory(y0, d, t, gamma):
    '''analytical solution for cumulative hitting time distribution'''
    numerator = sp.gammaincc((2 * gamma / d) + 1, (2 * y0) / (d * t))
    denominator = 1
    value = 1 - (numerator / denominator)
    return value
```

plotting cumulative hitting time against analytical result

```python
sorted_hitting = np.sort(hitting)
cumsum = np.arange(len(hitting))/len(hitting)

fig, ax = plt.subplots()
plt.plot(sorted_hitting, -1+cumsum, marker='o')
plt.plot(time, -cum_theory(2, D, time, gamma))

sns.despine()

ax.set_xlim(0,fit_until_this_t)
```

# Analytical 2d probability VS data

implementing functions for 2d probability distribution

```python
def probability_2D_single(y, y0, d, gamma, dt):
    if y < y0:
        numerator = (4**((d+gamma)/d) * np.exp(-(2*y)/(d*dt)) * (d*dt)**(-(2*(d+gamma))/d) * y**((2*gamma)/d + 1)) / sp.gamma((2*(d+gamma))/d)
    else:
        numerator = (4**((d+gamma)/d) * np.exp(-(2*y)/(d*dt)) * (d*dt)**(-(2*(d+gamma))/d) * y0**((2*gamma)/d + 1)) / sp.gamma((2*(d+gamma))/d)

    denominator = 1  -  sp.gammaincc((2*gamma)/d + 1, (2*y0)/(d*dt))

    return numerator / denominator
    

def probability_2D(Y, y0, d, gamma, T):
    arr = np.zeros((len(T), len(Y)))
    for j in np.arange(len(Y)):
        for i in np.arange(len(T)):
            y = Y[i]
            dt = T[j]
            
            arr[i,j] = probability_2D_single(y, y0, d, gamma, dt)

    return arr


def probability_2D_single_nonnorm(y, y0, d, gamma, dt):
    if y < y0:
        numerator = (4**((d+gamma)/d) * np.exp(-(2*y)/(d*dt)) * (d*dt)**(-(2*(d+gamma))/d) * y**((2*gamma)/d + 1)) / sp.gamma((2*(d+gamma))/d)
    else:
        numerator = (4**((d+gamma)/d) * np.exp(-(2*y)/(d*dt)) * (d*dt)**(-(2*(d+gamma))/d) * y0**((2*gamma)/d + 1)) / sp.gamma((2*(d+gamma))/d)

    denominator = 1  -  sp.gammaincc((2*gamma)/d + 1, (2*y0)/(d*dt))

    return numerator / 1

def probability_2D_nonnorm(Y, y0, d, gamma, T):
    arr = np.zeros((len(T), len(Y)))
    for j in np.arange(len(Y)):
        for i in np.arange(len(T)):
            y = Y[i]
            dt = T[j]
            
            arr[i,j] = probability_2D_single_nonnorm(y, y0, d, gamma, dt)

    return arr
```
